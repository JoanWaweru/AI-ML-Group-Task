```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
n = 1000
p = 0.5
x = list(range(0,n+1))
q = 1-p
```

```python
x
```

```python
y= stats.binom.pmf(x, n, p)
z = stats.poisson.pmf(x,n*p)
a = stats.norm.pdf(x,n*p,n*p*q)
```

```python
y
```

```python
z
```

```python
a
```

```python
n1 = 3
p1 = 0.5
x1 = 4
```

```python
stats.binom.pmf(x1, n1, p1)
```

```python
plt.plot(x,y,label="Binomial")
plt.plot(x,z,label="Poisson")
plt.plot(x,a,label="Normal")
plt.ylabel("Probability")
plt.xlabel("RV.X")
plt.legend()
```
